```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(0, figsize=(20, 20))
X_cols_tem = [c for c in X_cols if c not in ['compactness_int', 'compactness_total']]
sns.heatmap(df[X_cols_tem + y_cols].corr('pearson'), vmin=-1, vmax=1, cmap='coolwarm', annot=True, fmt='.2f')
plt.savefig("../../results/correlation_matrix.png")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, (ax, col) in enumerate(zip(axes, y_cols)):
    sns.histplot(df[col], ax=ax, bins=30, color='steelblue')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.savefig('../../results/hist_quarternions.png')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, (ax, col) in enumerate(zip(axes, ['yaw_deg', 'pitch_deg', 'roll_deg'])):
    sns.histplot(df[col], ax=ax, bins=30, color='steelblue')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.savefig('../../results/hist_euler_angles.png')
plt.show()
```
